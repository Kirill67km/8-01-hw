##Задание 3. Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
missings = df.isnull()
total_missings = missings.sum()
total_missings
```

```python
# В строке 132 dataframe присутствует пропуск в столбцах 'surgery' и 'outcome', 
# исходя из суммарного количества всех пропусков, то в данных столбцах они встречаются по 1 разу

column_surgery = 'surgery'
missings_surgery = df[df['surgery'].isnull()]

column_outcome = 'outcome'
missings_outcome = df[df['outcome'].isnull()]

missings_surgery 
```

```python
# С учетом того, что мода их равна 1.0 и не пристуствуют выбросы,
# то заполняем пустые значени модами столбцов

df['surgery'].fillna(df['surgery'].mode()[0], inplace=True)
df['outcome'].fillna(df['outcome'].mode()[0], inplace=True)
```

```python
# Поскольку в данных столбцах больше всего выбросов, то целесообразно выбрать медиану, так как
# она меня чувствительная к ним, также стоит обратить внимание на то, что необходимо также
# группировать данные в столбцах, иначе возможно не заполнение всех пропусков в данных

df['pain'].fillna(df['pain'].median(), inplace=True)
df['rectal temperature'].fillna(df.groupby(['surgery', 'age'])['rectal temperature'].transform('median'), inplace=True)
df['pulse'].fillna(df.groupby(['pain', 'age'])['pulse'].transform('median'), inplace=True)
df['temperature of extremities'].fillna(df.groupby(['age', 'outcome'])['temperature of extremities'].transform('median'), inplace=True)
df['respiratory rate'].fillna(df.groupby(['outcome', 'age'])['respiratory rate'].transform('median'), inplace=True)
df['respiratory rate'].fillna(df['respiratory rate'].median(), inplace=True)
```
